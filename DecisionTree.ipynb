```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score
from sklearn.model_selection import RandomizedSearchCV
```

```python
train = pd.read_csv("Train.csv")
```

```python
train.head()
```

```python
X = train.drop(columns="class")
y = train["class"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.3)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
Model = DecisionTreeClassifier(criterion='gini',max_depth=7)
Model.fit(X_train,y_train)
```

```python
y_pred = Model.predict(X_test)
```

```python
Model.score(X_train,y_train)
```

```python
f1_score(y_test,y_pred)
```

```python
mod= DecisionTreeClassifier()
```

```python
param_dist = {
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random'],
    'max_depth': [None,2,3,4,5,6,7,8,9, 10, 20, 30, 40, 50],
    'min_samples_split': [2,3,4,5,6,7,8,9,11,12,13, 10, 20],
    'min_samples_leaf': [1,2,3,4,6,7,8,9, 5, 10],
    
}

# Set up RandomizedSearchCV
random_search = RandomizedSearchCV(estimator=mod, param_distributions=param_dist, n_iter=200, cv=5, scoring='f1')

random_search.fit(X_train,y_train)
```

```python
y = random_search.predict(X_test)
```

```python
f1_score(y_test,y)
```

```python
test = pd.read_csv("Test.csv")
Index = test.Index
test.drop(columns="Index", inplace=True)
test_scaled = scaler.transform(test)
```

```python
y_pred = random_search.predict(test)
```

```python
result = pd.DataFrame({
    "Index" : Index,
    "class" : y_pred
})
result.to_csv("2035_Mithil 3.csv", index=False)
```
